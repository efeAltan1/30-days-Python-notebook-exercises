## Exercise 1 — Find All Numbers

Use `re.findall()` to extract all numbers from a string. Return them as a list of **integers**.

| Input | Expected Output |
|---|---|
| `'I have 2 cats and 3 dogs'` | `[2, 3]` |
| `'Room 101 is on floor 3'` | `[101, 3]` |
| `'No numbers here'` | `[]` |

````python
import re

def extract_numbers(text):
    matches = re.findall(r'\d+' , text)
    return [int(i) for i in matches]
````

````python
# TEST CELL — do not modify
def _check(label, got, expected):
    if got == expected:
        print(f'  ✔ APPROVED — {label}')
    else:
        print(f'  ✘ FAILED   — {label}')
        print(f'             got:      {got!r}')
        print(f'             expected: {expected!r}')

print('\n── Exercise 1: extract_numbers ──')
_check('cats and dogs',   extract_numbers('I have 2 cats and 3 dogs'), [2,3])
_check('Room 101 floor 3',extract_numbers('Room 101 is on floor 3'),   [101,3])
_check('no numbers',      extract_numbers('No numbers here'),           [])
````

## Exercise 2 — Validate Email

Use `re.match()` to check if a string is a valid email address.

A valid email:
- Has characters before `@`
- Has `@`
- Has a domain with a `.` and at least 2 letters after it

Return `True` if valid, `False` if not.

| Input | Expected Output |
|---|---|
| `'user@example.com'` | `True` |
| `'bad-email'` | `False` |
| `'missing@domain'` | `False` |
| `'a@b.co'` | `True` |

````python
import re

def is_valid_email(email):
    if re.match(r'\S+@\S+\.[a-zA-Z]{2,}', email) is not None:
        return True
    else:
        return False
````

````python
def _check(label, got, expected):
    if got == expected:
        print(f'  APPROVED - {label}')
    else:
        print(f'  FAILED   - {label}')
        print(f'             got:      {got!r}')
        print(f'             expected: {expected!r}')

print('\n-- Exercise 2: is_valid_email --')
_check('valid email',     is_valid_email('user@example.com'), True)
_check('no @ sign',       is_valid_email('bad-email'),        False)
_check('no domain ext',   is_valid_email('missing@domain'),   False)
_check('short but valid', is_valid_email('a@b.co'),           True)
````

## Mini-Project — Log Parser

You have a server log string. Use regex to extract structured data from it.

Each line looks like:
```
[2024-01-15 08:32:11] ERROR 404 /page-not-found
[2024-01-15 08:33:02] INFO 200 /home
[2024-01-15 08:34:55] ERROR 500 /api/data
```

Write a function that parses this log and returns a list of dicts:
```python
[{'timestamp': '2024-01-15 08:32:11', 'level': 'ERROR', 'code': 404, 'path': '/page-not-found'}, ...]
```

Also write a function that filters only ERROR entries from the parsed log.

````python
import re

def parse_log(log_text):
    # Return list of dicts with keys: timestamp, level, code, path
    parsed_text = re.findall(r'\[(.+?)\] (\w+) (\d+) (\S+)', log_text)
    result = []
    for (timestamp, level, code, path) in parsed_text:
        result.append({'timestamp': timestamp, 'level': level, 'code': int(code), 'path': path})
    return result
       

def get_errors(parsed_log):
    # Return only entries where level == 'ERROR'
    return [i for i in parsed_logs[i] if i['level'] == 'ERROR']
````
