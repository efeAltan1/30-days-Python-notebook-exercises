## Exercise 1 — Basic Try/Except

Write a function that safely converts a string to an integer.

- If the conversion works → return the integer
- If it fails → return `None`

| Input | Expected Output |
|---|---|
| `'42'` | `42` |
| `'hello'` | `None` |
| `'3.14'` | `None` |
| `''` | `None` |

```python
def safe_int(value):
    try:
        return int(value)
    except ValueError:
        return None
    except TypeError:
        return None
```

```python
# TEST CELL — do not modify
def _check(label, got, expected):
    if got == expected:
        print(f'  ✔ APPROVED — {label}')
    else:
        print(f'  ✘ FAILED   — {label}')
        print(f'             got:      {got!r}')
        print(f'             expected: {expected!r}')

print('\n── Exercise 1: safe_int ──')
_check("safe_int('42')",    safe_int('42'),    42)
_check("safe_int('hello')", safe_int('hello'), None)
_check("safe_int('3.14')",  safe_int('3.14'),  None)
_check("safe_int('')",      safe_int(''),      None)
```

## Exercise 2 — Multiple Exception Types

Write a function that takes a list and an index, and safely retrieves the item at that index.

- If the index is out of range → return `'IndexError'`
- If the index is not an integer → return `'TypeError'`
- Otherwise → return the item

| Input | Expected Output |
|---|---|
| `safe_get([1,2,3], 1)` | `2` |
| `safe_get([1,2,3], 10)` | `'IndexError'` |
| `safe_get([1,2,3], 'a')` | `'TypeError'` |

```python
def safe_get(lst, index):
    try:
        return lst[index]
    except IndexError:
        return 'IndexError'
    except TypeError:
        return 'TypeError'
```

```python
def _check(label, got, expected):
    if got == expected:
        print(f'  APPROVED - {label}')
    else:
        print(f'  FAILED   - {label}')
        print(f'             got:      {got!r}')
        print(f'             expected: {expected!r}')

print('\n-- Exercise 2: safe_get --')
_check('valid index',   safe_get([1,2,3], 1),   2)
_check('out of range',  safe_get([1,2,3], 10),  'IndexError')
_check('wrong type',    safe_get([1,2,3], 'a'), 'TypeError')
```

## Mini-Project — Input Validator

Write a function that validates and processes user registration data.

Input is a dict. Validate these rules **in order** and raise the appropriate exception if a rule fails:

1. `'name'` must be a non-empty string → raise `ValueError('Name cannot be empty')`
2. `'age'` must be an integer between 0–120 → raise `ValueError('Age must be between 0 and 120')`
3. `'email'` must contain `'@'` and `'.'` → raise `ValueError('Invalid email format')`

If all valid → return `{'status': 'ok', 'name': name, 'age': age, 'email': email}`

Catch the `ValueError` in a wrapper and return `{'status': 'error', 'message': str(e)}`

```python
def validate_user(data):
    try:
        if data['name'] == '':
            raise ValueError('Name cannot be empty')
        elif not 0 <= data['age'] <= 120:
            raise ValueError('Age must be between 0 and 120')
        elif '@' not in data['email'] or '.' not in data['email']:
            raise ValueError('Invalid email format')
        
        return {'status': 'ok', 'name': data['name'], 'age': data['age'], 'email': data['email']}

    except ValueError as e:
        return {'status': 'error', 'message': str(e)}
```

```python
# TEST CELL — do not modify
def _check(label, got, expected):
    if got == expected:
        print(f'  ✔ APPROVED — {label}')
    else:
        print(f'  ✘ FAILED   — {label}')
        print(f'             got:      {got!r}')
        print(f'             expected: {expected!r}')

print('\n── Mini-Project: validate_user ──')
_check('valid user', validate_user({'name':'Alice','age':25,'email':'alice@mail.com'}),
       {'status':'ok','name':'Alice','age':25,'email':'alice@mail.com'})
_check('empty name', validate_user({'name':'','age':25,'email':'a@b.com'}),
       {'status':'error','message':'Name cannot be empty'})
_check('bad age',    validate_user({'name':'Bob','age':200,'email':'b@b.com'}),
       {'status':'error','message':'Age must be between 0 and 120'})
_check('bad email',  validate_user({'name':'Carol','age':30,'email':'notanemail'}),
       {'status':'error','message':'Invalid email format'})
```
