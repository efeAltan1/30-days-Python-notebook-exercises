## Exercise 2 — Days Between Dates

Write a function that takes two date strings in `'YYYY-MM-DD'` format and returns the number of days between them as a positive integer.

**Hint:** Use `strptime()` to parse the strings, then subtract.

| Input | Expected Output |
|---|---|
| `days_between('2024-01-01', '2024-03-15')` | `74` |
| `days_between('2023-06-10', '2023-01-01')` | `160` |

```python
from datetime import datetime

def days_between(date1_str, date2_str):
    t1 = datetime.strptime(date1_str, '%Y-%m-%d')
    t2 = datetime.strptime(date2_str, '%Y-%m-%d')
    diff = t2 - t1
    return abs(diff.days)
```

## Mini-Project — Age Calculator

Write a function that takes a birthdate string (`"YYYY-MM-DD"`) and today's date, and returns:

- `'age'`: full years old
- `'next_birthday_in'`: days until next birthday (from today)
- `'day_of_week_born'`: day of the week they were born (e.g. `'Monday'`)

**Hint:** Use `timedelta`, `strftime('%A')` for day name, and be careful with leap year birthdays.

```python
from datetime import datetime, date, timedelta

def age_info(birthdate_str, today_str):
    # birthdate_str and today_str format: 'YYYY-MM-DD'
    t1 = datetime.strptime(birthdate_str, '%Y-%m-%d')
    t2 = datetime.strptime(today_str, '%Y-%m-%d')
    age = (t2 - t1).days / 365

    next_birthday = datetime(t2.year, t1.month, t1.day)
    if next_birthday < t2:
        next_birthday = datetime(t2.year + 1, t1.month, t1.day)
    days_until = (next_birthday - t2).days
    

    return{
        'age': int(age),
        'next_birthday_in': abs(days_until),
        'day_of_week_born': t1.strftime('%A')
    }
```
