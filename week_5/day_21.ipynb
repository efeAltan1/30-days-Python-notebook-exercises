## Exercise 1 — Person Class

Create a `Person` class with:
- `__init__(self, name, age, country)` — stores all three as instance attributes
- `greet()` — returns `"Hi, I am {name} and I am {age} years old. I live in {country}."`
- `is_adult()` — returns `True` if age >= 18, else `False`
- `birthday()` — increments age by 1, returns the new age

```python
class Person:
    def __init__(self, name, age, country):
        self.name = name
        self.age = age
        self.country = country

    def greet(self):
        return f'Hi, I am {self.name} and I am {self.age} years old. I live in {self.country}.'

    def is_adult(self):
        if self.age >= 18:
            return True
        else:
            return False

    def birthday(self):
        for i in self.age:
            i += 1
            return self.age
    
```

## Exercise 2 — Animal Inheritance

Create a base class `Animal` and two subclasses `Dog` and `Cat`.

- `Animal.__init__(self, name, sound)` — stores both attributes
- `Animal.speak()` — returns `"{name} says {sound}!"`
- `Animal.describe()` — returns `"I am an animal named {name}."`
- `Dog` inherits from `Animal`. Its `__init__` only takes `name` and calls
  `super().__init__(name, 'Woof')`.
  - Override `describe()` to return `"I am a dog named {name}."`.
- `Cat` inherits from `Animal`. Its `__init__` only takes `name` and calls
  `super().__init__(name, 'Meow')`.
  - Override `describe()` to return `"I am a cat named {name}."`.

```python
class Animal:
    def __init__(self, name, sound):
        self.name = name
        self.sound = sound

    def speak(self):
        return f'{self.name} says {self.sound}!'

    def describe(self):
        return f'I am an animal named {self.name}.'

class Dog(Animal):
    def __init__(self, name):
        self.name = name
        super().__init__(name, 'Woof')
        
    def describe(self):
        return f'I am a dog named {self.name}.'

class Cat(Animal):
    def __init__(self, name):
        self.name = name
        super().__init__(name, 'Meow')

    def describe(self):
        return f'I am a cat named {self.name}.'
```

```python
# TEST CELL — do not modify
def _check(label, got, expected):
    if got == expected:
        print(f'  ✔ APPROVED — {label}')
    else:
        print(f'  ✘ FAILED   — {label}')
        print(f'             got:      {got!r}')
        print(f'             expected: {expected!r}')

print('\n── Exercise 2: Animal Inheritance ──')
a = Animal('Rex', 'Roar')
_check('Animal.speak',    a.speak(),    'Rex says Roar!')
_check('Animal.describe', a.describe(), 'I am an animal named Rex.')

d = Dog('Buddy')
_check('Dog.speak',       d.speak(),    'Buddy says Woof!')
_check('Dog.describe',    d.describe(), 'I am a dog named Buddy.')

c = Cat('Whiskers')
_check('Cat.speak',       c.speak(),    'Whiskers says Meow!')
_check('Cat.describe',    c.describe(), 'I am a cat named Whiskers.')
_check('Dog is Animal',   isinstance(d, Animal), True)
_check('Cat is Animal',   isinstance(c, Animal), True)
```

## Mini-Project — Bank Account

Build a `BankAccount` class that manages deposits, withdrawals, and transfers.

- `__init__(self, owner, balance=0)`
- `deposit(amount)` — adds to balance. Raises `ValueError` if amount <= 0. Returns new balance.
- `withdraw(amount)` — subtracts from balance. Raises `ValueError` if amount > balance or amount <= 0. Returns new balance.
- `get_balance()` — returns current balance
- `transfer(other_account, amount)` — withdraws from self, deposits to other. Returns tuple `(self_balance, other_balance)`.

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError('ValueError')
        else:
            self.balance = self.balance + amount
            return self.balance

    def withdraw(self, amount):
        if amount > self.balance or amount <= 0:
            raise ValueError('ValueError')
        else:
            self.balance = self.balance - amount
            return self.balance

    def get_balance(self):
        return self.balance

    def transfer(self, other_account, amount):
        self.withdraw(amount)
        other_account.deposit(amount)
        return (self.balance, other_account.balance)
```

```python
# TEST CELL — do not modify
def _check(label, got, expected):
    if got == expected:
        print(f'  ✔ APPROVED — {label}')
    else:
        print(f'  ✘ FAILED   — {label}')
        print(f'             got:      {got!r}')
        print(f'             expected: {expected!r}')

print('\n── Mini-Project: BankAccount ──')
acc = BankAccount('Alice', 100)
_check('initial balance', acc.get_balance(), 100)
_check('deposit 50',      acc.deposit(50),   150)
_check('withdraw 30',     acc.withdraw(30),  120)

try:
    acc.withdraw(9999)
    _check('overdraft raises', False, True)
except ValueError:
    _check('overdraft raises ValueError', True, True)

try:
    acc.deposit(-10)
    _check('negative deposit raises', False, True)
except ValueError:
    _check('negative deposit raises ValueError', True, True)

other = BankAccount('Bob', 0)
self_bal, other_bal = acc.transfer(other, 50)
_check('transfer self balance',  self_bal,  70)
_check('transfer other balance', other_bal, 50)
```
