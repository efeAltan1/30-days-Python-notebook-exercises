## Exercise 1 — Array Creation and Basic Math

Using NumPy, write a function that:
1. Creates a NumPy array from the input list
2. Returns a dict with element-wise results:
   - `'doubled'`: every element × 2
   - `'squared'`: every element ² 
   - `'normalized'`: every element divided by the max value (as float, rounded to 2 decimals each)

Return all values as regular Python lists (use `.tolist()`).

````python
import numpy as np

def array_math(numbers):
    array_numbers = np.array(numbers)
    array_doubled = (array_numbers * 2).tolist()
    array_squared = (array_numbers ** 2).tolist()
    array_normalized = array_numbers / max(array_numbers)
    return{
        'doubled': array_doubled,
        'squared': array_squared,
        'normalized': np.round(array_normalized, 2).tolist()
    }
    
    
````

## Exercise 2 — Matrix Statistics

Write a function `matrix_stats(matrix)` that accepts a 2-D list and uses NumPy to return a dict:
- `'shape'`: tuple `(rows, cols)` — use regular Python tuple
- `'row_sums'`: list of sums per row (as ints or floats)
- `'col_means'`: list of means per column, each rounded to 2 decimals
- `'max_value'`: single maximum value across the whole matrix
- `'flattened'`: flattened list of all values in row-major order

````python
import numpy as np

def matrix_stats(matrix):
    array_matrix = np.array(matrix)
    array_shape = array_matrix.shape
    array_row_sums = np.sum(array_matrix, axis=1).tolist()
    array_col_means = np.mean(array_matrix, axis=0).tolist()
    array_max_value = np.max(array_matrix).tolist()
    array_flattened = array_matrix.flatten().tolist()

    return{
        'shape': array_shape,
        'row_sums': array_row_sums,
        'col_means': array_col_means,
        'max_value': array_max_value,
        'flattened': array_flattened
    }

````

````python
# TEST CELL — do not modify
def _check(label, got, expected):
    if got == expected:
        print(f'  ✔ APPROVED — {label}')
    else:
        print(f'  ✘ FAILED   — {label}')
        print(f'             got:      {got!r}')
        print(f'             expected: {expected!r}')

print('\n── Exercise 2: matrix_stats ──')
m = [[1, 2, 3],
     [4, 5, 6],
     [7, 8, 9]]
result = matrix_stats(m)
_check('shape',     result.get('shape'),     (3, 3))
_check('row_sums',  result.get('row_sums'),  [6, 15, 24])
_check('col_means', result.get('col_means'), [4.0, 5.0, 6.0])
_check('max_value', result.get('max_value'), 9)
_check('flattened', result.get('flattened'), [1,2,3,4,5,6,7,8,9])
````

## Mini-Project — Grade Analyzer with NumPy

You have a class of students with scores in three subjects. Use NumPy to analyze the results.

Input: a 2D list where each row is a student's scores `[math, science, english]`

Return a dict:
- `'class_average'`: overall mean across all scores (rounded to 2 decimals)
- `'subject_averages'`: list of means per subject `[math_avg, sci_avg, eng_avg]` (rounded to 2 decimals)
- `'top_student_index'`: index of student with highest total score
- `'passing_students'`: count of students whose average score >= 60

````python
import numpy as np

def grade_analyzer(scores):
    array_scores = np.array(scores, dtype=float)
    array_class_avg = float(np.mean(array_scores))
    array_subject_avg = np.mean(array_scores, axis=0).tolist()
    array_top_student_index = np.argmax(np.sum(array_scores, axis=1).tolist())
    row_means = np.mean(array_scores, axis=1)
    passing_students = int((row_means >= 60).sum())

    return{
        'class_average': round(array_class_avg, 2),
        'subject_averages': np.round(array_subject_avg, 2).tolist(),
        'top_student_index': array_top_student_index,
        'passing_students': passing_students
    }
````

````python
# TEST CELL — do not modify
def _check(label, got, expected):
    if got == expected:
        print(f'  ✔ APPROVED — {label}')
    else:
        print(f'  ✘ FAILED   — {label}')
        print(f'             got:      {got!r}')
        print(f'             expected: {expected!r}')

scores = [
    [90, 85, 92],   # student 0 — avg 89
    [55, 60, 50],   # student 1 — avg 55
    [78, 82, 80],   # student 2 — avg 80
    [40, 45, 35],   # student 3 — avg 40
]
print('\n── Mini-Project: grade_analyzer ──')
result = grade_analyzer(scores)
_check('class_average',     result.get('class_average'),     66.0)
_check('subject_averages',  result.get('subject_averages'),  [65.75,68.0,64.25])
_check('top_student_index', result.get('top_student_index'), 0)
_check('passing_students',  result.get('passing_students'),  2)
````
