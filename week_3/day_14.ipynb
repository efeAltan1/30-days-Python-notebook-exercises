## Exercise 1 — `map()`

Use `map()` to apply a transformation to every item in a list.

Write a function that takes a list of temperatures in Celsius and returns them converted to Fahrenheit.

Formula: `F = (C * 9/5) + 32`

⚠️ You must use `map()` — no list comprehensions or loops.

Return a **list** (not a map object). Round each value to 1 decimal.

| Input | Expected Output |
|---|---|
| `[0, 100, 37]` | `[32.0, 212.0, 98.6]` |
| `[-40, 20]` | `[-40.0, 68.0]` |

````python
def celsius_to_fahrenheit(celsius_list):
    conv_list = map(lambda i: round((i * 9/5 + 32), 1), celsius_list)
    return list(conv_list)
````

````python
# TEST CELL — do not modify
def _check(label, got, expected):
    if got == expected:
        print(f'  ✔ APPROVED — {label}')
    else:
        print(f'  ✘ FAILED   — {label}')
        print(f'             got:      {got!r}')
        print(f'             expected: {expected!r}')

print('\n── Exercise 1: celsius_to_fahrenheit ──')
_check('[0,100,37]', celsius_to_fahrenheit([0,100,37]), [32.0,212.0,98.6])
_check('[-40,20]',   celsius_to_fahrenheit([-40,20]),   [-40.0,68.0])
````

## Exercise 2 — Closure

A **closure** is a function that remembers a value from its enclosing scope.

Write a function `make_multiplier(n)` that **returns a new function**. The returned function takes a number and multiplies it by `n`.

```python
double = make_multiplier(2)
double(5)  # → 10

triple = make_multiplier(3)
triple(5)  # → 15
```

````python
def make_multiplier(n):
    def multiply(x):
        return n * x
    
    return multiply


````

````python
# TEST CELL — do not modify
def _check(label, got, expected):
    if got == expected:
        print(f'  ✔ APPROVED — {label}')
    else:
        print(f'  ✘ FAILED   — {label}')
        print(f'             got:      {got!r}')
        print(f'             expected: {expected!r}')

print('\n── Exercise 2: make_multiplier ──')
double = make_multiplier(2)
triple = make_multiplier(3)
_check('double(5)',  double(5),  10)
_check('double(0)',  double(0),  0)
_check('triple(5)',  triple(5),  15)
_check('triple(7)',  triple(7),  21)
````

## Mini-Project — Data Pipeline

Chain `map()`, `filter()`, and `reduce()` together to process a list of student scores.

Given a list of raw scores (0–100):
1. **Filter** out scores below 50 (failing)
2. **Map** each passing score to a letter grade (`A`=90+, `B`=80+, `C`=70+, `D`=50+)
3. **Reduce** to count how many of each grade appear — return a dict

Return: `{'A': n, 'B': n, 'C': n, 'D': n}` (only include grades that appear)

````python
from functools import reduce

def grade_pipeline(scores):
    # Step 1: keep only scores >= 50
    passing = filter(lambda i: i >= 50, scores)
    
    # Step 2: convert each score to a letter grade
    grades = map(lambda i: 'A' if i >= 90 else 'B' if i >= 80 else 'C' if i >= 70 else 'D', passing)
    
    # Step 3: count each grade into a dict
    def count_grades(acc, grade):
        acc[grade] = acc.get(grade, 0) + 1
        return acc
    
    return reduce(count_grades, grades, {})




    
````

````python
# TEST CELL — do not modify
def _check(label, got, expected):
    if got == expected:
        print(f'  ✔ APPROVED — {label}')
    else:
        print(f'  ✘ FAILED   — {label}')
        print(f'             got:      {got!r}')
        print(f'             expected: {expected!r}')

print('\n── Mini-Project: grade_pipeline ──')
_check('mixed scores', grade_pipeline([95,82,45,71,88,55,91,30,66]),
       {'A':2,'B':2,'C':1,'D':2})
_check('all failing',  grade_pipeline([10,20,30,40]), {})
````
