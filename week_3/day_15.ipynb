## Exercise 1 — Identify the Error

Write a function that takes a description and returns the correct error type **as a string**.

| Description | Expected Answer |
|---|---|
| `'adding a string to an integer'` | `'TypeError'` |
| `'dividing a number by zero'` | `'ZeroDivisionError'` |
| `'accessing index 10 on a 3-item list'` | `'IndexError'` |
| `'using a key that does not exist in a dict'` | `'KeyError'` |
| `'converting the string "hello" to int'` | `'ValueError'` |
| `'using a variable before defining it'` | `'NameError'` |

```python
def identify_error(description):
    # Map each description to the correct error type string

    errors = {
        'adding a string to an integer': 'TypeError',
        'dividing a number by zero': 'ZeroDivisionError',
        'accessing index 10 on a 3-item list': 'IndexError',
        'using a key that does not exist in a dict': 'KeyError',
        'converting the string "hello" to int': 'ValueError',
        'using a variable before defining it': 'NameError'

    }
    return errors[description]
```

```python
# TEST CELL — do not modify
def _check(label, got, expected):
    if got == expected:
        print(f'  ✔ APPROVED — {label}')
    else:
        print(f'  ✘ FAILED   — {label}')
        print(f'             got:      {got!r}')
        print(f'             expected: {expected!r}')

print('\n── Exercise 1: identify_error ──')
_check('str + int',        identify_error('adding a string to an integer'),          'TypeError')
_check('divide by zero',   identify_error('dividing a number by zero'),               'ZeroDivisionError')
_check('bad index',        identify_error('accessing index 10 on a 3-item list'),     'IndexError')
_check('bad dict key',     identify_error('using a key that does not exist in a dict'), 'KeyError')
_check('bad int cast',     identify_error('converting the string "hello" to int'),    'ValueError')
_check('undefined var',    identify_error('using a variable before defining it'),     'NameError')
```

## Exercise 2 - Safe Calculator

Write a function that performs a calculation but handles errors gracefully using try/except.

It takes two values a, b and an operator op ('+', '-', '*', '/').

| Situation | Return |
|---|---|
| Normal result | the calculated value |
| Division by zero | 'ZeroDivisionError' |
| Invalid types | 'TypeError' |
| Unknown operator | 'ValueError' |

| Input | Expected |
|---|---|
| safe_calc(10, 2, '/') | 5.0 |
| safe_calc(10, 0, '/') | 'ZeroDivisionError' |
| safe_calc('a', 2, '+') | 'TypeError' |
| safe_calc(10, 2, '?') | 'ValueError' |

```python
def safe_calc(a, b, op):
    try:
        if op == '+':
            return a + b
        elif op == '-':
            return a - b
        elif op == '/':
            return a / b
        elif op == '*':
            return a * b
        else: 
            raise ValueError

    except ZeroDivisionError:
        return 'ZeroDivisionError'
    except TypeError:
        return 'TypeError'
    except ValueError:
        return 'ValueError'
    

    
```

```python
def _check(label, got, expected):
    if got == expected:
        print(f'  APPROVED - {label}')
    else:
        print(f'  FAILED   - {label}')
        print(f'             got:      {got!r}')
        print(f'             expected: {expected!r}')

print('-- Exercise 2: safe_calc --')
_check('10 / 2',   safe_calc(10, 2,   '/'), 5.0)
_check('10 / 0',   safe_calc(10, 0,   '/'), 'ZeroDivisionError')
_check('a + 2',    safe_calc('a', 2,  '+'), 'TypeError')
_check('10 ? 2',   safe_calc(10, 2,   '?'), 'ValueError')
```

## Week 3 Capstone - Student Grade Manager

This capstone combines functions, modules, list comprehensions, higher-order functions, and error handling from Days 11-15.

Student data is stored as a list of dicts: [{'name': str, 'grades': [int]}]

---
### add_student(students, name, grades)
Add a student. If name already exists raise ValueError('Student already exists').

### get_average(students, name)
Return the student's average grade rounded to 2 decimals.
Raise KeyError if student not found.

### top_students(students, n)
Return a sorted list of the top n student names by average grade (highest first).

### grade_distribution(students)
Return a dict counting how many students fall in each band:
{'A': count, 'B': count, 'C': count, 'D': count, 'F': count}
(A=90+, B=80-89, C=70-79, D=60-69, F=below 60)

### failing_students(students)
Return a sorted list of names of students whose average is below 60.

```python
def add_student(students, name, grades):
    if any(s['name'] == name for s in students):
        raise ValueError('Student already exists')
    else:
        students.append({'name': name, 'grades': grades})
        
    return students
    
def get_average(students, name):
    for student in students:
        if student['name'] == name:
            return round(sum(student['grades']) / len(student['grades']), 2)
    raise KeyError('Student not found')

def top_students(students, n):
    sort_top = sorted(students, key=lambda student: get_average(students, student['name']), reverse=True)
    first_name = sort_top[:n]
    return [student['name'] for student in first_name]

def grade_distribution(students):
    results = {}
    for i in students:
        avg = get_average(students, i['name'])
        if avg >= 90:
            grade = 'A'
        elif avg >= 80:
            grade = 'B'
        elif avg >= 70:
            grade = 'C'
        elif avg >= 60:
            grade = 'D'
        else:
            grade = 'F'

        results[grade] = results.get(grade, 0) + 1

    return results
    
def failing_students(students):
    failing = []
    
    for i in students:
        avg = get_average(students, i['name'])
        if avg < 60:
            failing.append(i['name'])

    return sorted(failing)
```

```python
def _check(label, got, expected):
    if got == expected:
        print(f'  APPROVED - {label}')
    else:
        print(f'  FAILED   - {label}')
        print(f'             got:      {got!r}')
        print(f'             expected: {expected!r}')

print('-- Week 3 Capstone: Grade Manager --')
db = []
db = add_student(db, 'Alice', [92, 88, 95])
db = add_student(db, 'Bob',   [55, 60, 58])
db = add_student(db, 'Eve',   [78, 82, 80])
db = add_student(db, 'Dan',   [91, 94, 97])
_check('Alice avg',      get_average(db, 'Alice'), 91.67)
_check('Bob avg',        get_average(db, 'Bob'),   57.67)
_check('top 2',          top_students(db, 2),      ['Dan', 'Alice'])
_check('failing',        failing_students(db),     ['Bob'])
_check('dist A',         grade_distribution(db).get('A'), 2)
_check('dist B',         grade_distribution(db).get('B'), 1)
_check('dist F',         grade_distribution(db).get('F'), 1)
try:
    add_student(db, 'Alice', [99])
    _check('duplicate raises', False, True)
except ValueError:
    _check('duplicate raises ValueError', True, True)
try:
    get_average(db, 'Ghost')
    _check('missing raises', False, True)
except KeyError:
    _check('missing raises KeyError', True, True)
```
