## Exercise 1 — The `math` Module

Use the `math` module to complete this function. Return a dict with:

| Key | Value |
|---|---|
| `'sqrt'` | square root of n (rounded to 2 decimals) |
| `'ceil'` | ceiling of n |
| `'floor'` | floor of n |
| `'is_whole'` | True if n has no decimal part |

**Hint:** `import math` at the top, then use `math.sqrt()`, `math.ceil()`, `math.floor()`

```python
import math as m

def math_ops(n):
    return{

        'sqrt': round(m.sqrt(n), 2),
        'ceil': m.ceil(n),
        'floor': m.floor(n),
        'is_whole': n % 1 == 0

    }
    
```

## Exercise 2 - The statistics Module

Use the built-in statistics module to analyse a list of numbers.

Return a dictionary with:

| Key | Value |
|---|---|
| 'mean' | average (rounded to 2 decimals) |
| 'median' | middle value |
| 'mode' | most common value |
| 'stdev' | standard deviation (rounded to 2 decimals) |

Hint: import statistics, then use statistics.mean(), statistics.median(), statistics.mode(), statistics.stdev()

```python
import statistics as s

def stats_summary(numbers):
    return{
        'mean': round(s.mean(numbers),2),
        'median': s.median(numbers),
        'mode': s.mode(numbers),
        'stdev': round(s.stdev(numbers))
    }
```

```python
def _check(label, got, expected):
    if got == expected:
        print(f'  APPROVED - {label}')
    else:
        print(f'  FAILED   - {label}')
        print(f'             got:      {got!r}')
        print(f'             expected: {expected!r}')

print('\n-- Exercise 2: stats_summary --')
r = stats_summary([2, 4, 4, 4, 5, 5, 7, 9])
_check('mean',   r.get('mean'),   5.0)
_check('median', r.get('median'), 4.5)
_check('mode',   r.get('mode'),   4)
_check('stdev',  r.get('stdev'),  2.0)
```

## Mini-Project — Password Generator

Use the `string` and `random` modules to generate a random password.

**Rules:**
- Length is specified by the `length` parameter
- Must include at least one uppercase letter, one lowercase letter, one digit, and one punctuation character
- The rest is filled randomly from all character types combined

Return the password as a string.

Since output is random, we test that:
1. Length matches
2. Has at least one uppercase
3. Has at least one lowercase
4. Has at least one digit
5. Has at least one punctuation

**Hint:** `string.ascii_uppercase`, `string.ascii_lowercase`, `string.digits`, `string.punctuation`

```python
import random as r
import string as s

def generate_password(length=12):
    password = []
    allowed_chars = s.ascii_uppercase + s.ascii_lowercase + s.digits + s.punctuation

    password.append(r.choice(s.ascii_uppercase))
    password.append(r.choice(s.ascii_lowercase))
    password.append(r.choice(s.digits))
    password.append(r.choice(s.punctuation))

    for i in range(length - 4):
        password.append(r.choice(allowed_chars))
        i += 1

    r.shuffle(password)
    return ''.join(password)
        
        
```

```python
# TEST CELL — do not modify
import string

def _check(label, got, expected):
    if got == expected:
        print(f'  ✔ APPROVED — {label}')
    else:
        print(f'  ✘ FAILED   — {label}')
        print(f'             got:      {got!r}')
        print(f'             expected: {expected!r}')

print('\n── Mini-Project: generate_password ──')
for length in [8, 12, 16]:
    pwd = generate_password(length)
    _check(f'length={length} correct',   len(pwd) == length, True)
    _check(f'length={length} uppercase', any(c in string.ascii_uppercase for c in pwd), True)
    _check(f'length={length} lowercase', any(c in string.ascii_lowercase for c in pwd), True)
    _check(f'length={length} digit',     any(c in string.digits for c in pwd), True)
    _check(f'length={length} punct',     any(c in string.punctuation for c in pwd), True)
```
