## Exercise 2 — URL Query String Parser

In Flask you often parse query strings from URLs. Simulate this without Flask.

Write `parse_query_string(query)` that:
- Accepts a raw query string like `"name=Alice&age=30&city=Istanbul"`
- Returns a dict mapping each key to its value **as a string**
- If the same key appears multiple times, the **last** value wins
- Empty string `""` → return `{}`

````python
def parse_query_string(query):
    results = {}
    query_split_and = query.split('&')
    if query == "":
        return {}
    
    for i in query_split_and:
        query_split_equal = i.split('=')
        results[query_split_equal[0]] = query_split_equal[1]

    return results
    
````

````python
# TEST CELL — do not modify
def _check(label, got, expected):
    if got == expected:
        print(f'  ✔ APPROVED — {label}')
    else:
        print(f'  ✘ FAILED   — {label}')
        print(f'             got:      {got!r}')
        print(f'             expected: {expected!r}')

print('\n── Exercise 2: parse_query_string ──')
_check('basic',    parse_query_string('name=Alice&age=30&city=Istanbul'),
                   {'name': 'Alice', 'age': '30', 'city': 'Istanbul'})
_check('empty',    parse_query_string(''),  {})
_check('single',   parse_query_string('token=abc123'), {'token': 'abc123'})
_check('duplicate last wins',
       parse_query_string('color=red&size=M&color=blue'),
       {'color': 'blue', 'size': 'M'})
````

## Mini-Project — Mini Flask App Logic

Build the complete logic layer for a simple blog API. No Flask needed — just the pure functions that route handlers would call.

```python
# In-memory store
posts = []

get_all_posts()         # returns all posts
create_post(title, body, author)  # adds post with auto-incrementing id, returns it
get_post(post_id)       # returns post or {'error': 'Not found'}
delete_post(post_id)    # removes post, returns {'deleted': True} or {'error': 'Not found'}
```

````python
posts = []

def get_all_posts():
    return posts

def create_post(title, body, author):
    post_dict = {'title': title, 'body': body, 'author':author, 'id': len(posts) + 1}
    posts.append(post_dict)
    return post_dict

def get_post(post_id):
    for post in posts:
        if post['id'] == post[id]:
            return post
        
    return {'error': 'Not found'}

def delete_post(post_id):
    for post in posts:
        if post['id'] == post[id]:
            posts.remove(post)
            return {'deleted': True}
        
    return {'error': 'Not found'}
````
