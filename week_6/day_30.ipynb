## What is SMP?

SMP (Student Management Platform) is a personal academic tracking system. It lets a student:

- Track **subjects** with teacher info and credit hours
- Log **grades** per subject with GPA calculation
- Manage **assignment deadlines** with priority and status
- Record **upcoming exam dates** with location and duration
- Track **attendance** per subject per day
- Store **announcements** and personal notes

It is built as a **full-stack web application**:
- A **Python/Flask REST API** as the backend
- **MongoDB Atlas** as the cloud database
- A **React** frontend as the UI
- Deployed on **Render** (backend) and **Vercel** (frontend)

## MongoDB Schema Design

All `_id` fields are auto-generated by MongoDB (`ObjectId`). All `subject_id` fields reference a document in the `subjects` collection.

---

### Collection: `subjects`
```json
{
  "_id": "ObjectId",
  "name": "Mathematics",
  "code": "MATH101",
  "teacher": "Dr. Smith",
  "credits": 4,
  "color": "#4A90E2",
  "created_at": "2024-01-15T08:00:00"
}
```

---

### Collection: `grades`
```json
{
  "_id": "ObjectId",
  "subject_id": "ObjectId",
  "subject_name": "Mathematics",
  "type": "Midterm",
  "score": 87.5,
  "max_score": 100,
  "weight": 0.3,
  "date": "2024-03-15",
  "notes": "Struggled with integration"
}
```
> GPA is always **calculated on the fly** using NumPy and Pandas -- never stored.

> `status` values: `pending`, `in_progress`, `done`

> `status` values: `present`, `absent`, `late`

## Model Classes (backend/models/)

All model classes live in `backend/models/`. They handle validation, serialization, and business logic. Flask routes call these classes -- routes never talk directly to MongoDB.

---

### `BaseModel` (base.py)
Parent class inherited by all models. Applies OOP concepts from Days 21-22.

**Attributes:**
- `collection_name` -- name of the MongoDB collection (overridden by each subclass)
- `db` -- pymongo database reference

**Methods:**
- `to_dict()` -- serializes the object to a JSON-safe dict
- `validate()` -- raises `ValueError` if required fields are missing (Day 17)
- `save()` -- inserts the document into MongoDB (Day 27)
- `@staticmethod find_by_id(id)` -- fetches one document by `_id`
- `@staticmethod find_all()` -- fetches all documents in the collection

---

### `Subject` (subject.py)
Inherits `BaseModel`. Represents one course.

**Attributes:** `name`, `code`, `teacher`, `credits`, `color`, `created_at`

**Methods:**
- `validate()` -- checks name/code are non-empty; validates code format with regex `r'^[A-Z]{2,10}\d{3}$'` (Day 18)
- `get_grade_summary()` -- uses Pandas to return average score per grade type (Day 25)

---

### `Grade` (grade.py)
Inherits `BaseModel`. Represents one graded assessment.

**Attributes:** `subject_id`, `subject_name`, `type`, `score`, `max_score`, `weight`, `date`, `notes`

**Methods:**
- `validate()` -- checks score is between 0 and max_score, weight between 0 and 1 (Day 17)
- `percentage()` -- returns `score / max_score * 100` rounded to 2 decimals
- `@staticmethod calculate_gpa(grades_list)` -- uses NumPy weighted average, maps to 4.0 scale (Day 24)

**GPA Scale:**
```
90-100  ->  4.0
80-89   ->  3.0
70-79   ->  2.0
60-69   ->  1.0
< 60    ->  0.0
```

---

### `Assignment` (assignment.py)
Inherits `BaseModel`.

**Attributes:** `subject_id`, `title`, `description`, `deadline`, `status`, `priority`, `created_at`

**Methods:**
- `validate()` -- checks deadline is a valid date string using `strptime` (Day 16), status is one of the allowed values
- `is_overdue()` -- compares deadline to `datetime.now()`, returns `True/False` (Day 16)
- `days_until_deadline()` -- returns integer days remaining; negative if overdue (Day 16)

---

### `Exam` (exam.py)
Inherits `BaseModel`.

**Attributes:** `subject_id`, `subject_name`, `type`, `date`, `location`, `duration_minutes`, `notes`

**Methods:**
- `validate()` -- date format validation with regex `r'\d{4}-\d{2}-\d{2}'` (Day 18)
- `@staticmethod upcoming(days=7)` -- returns exams within the next N days using `timedelta` (Day 16)

---

### `Attendance` (attendance.py)
Inherits `BaseModel`.

**Attributes:** `subject_id`, `date`, `status`, `notes`

**Methods:**
- `validate()` -- status must be one of `present`, `absent`, `late`
- `@staticmethod attendance_rate(subject_id)` -- fetches all records for a subject, returns percentage present using list comprehensions (Day 10)

---

### `Announcement` (announcement.py)
Inherits `BaseModel`.

**Attributes:** `title`, `body`, `subject_id`, `priority`, `created_at`, `tags`

**Methods:**
- `validate()` -- title and body must be non-empty
- `@staticmethod filter_by_tag(tag)` -- returns announcements containing the given tag (Day 10 higher-order functions)
