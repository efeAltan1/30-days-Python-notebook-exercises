## Exercise 1 вАФ Circle Geometry

Use arithmetic operators to calculate a circle's **area** and **perimeter**.

- `PI = 3.14159`
- `area = PI * r¬≤`
- `perimeter = 2 * PI * r`

Return a tuple `(area, perimeter)`, both **rounded to 2 decimal places**.

| Input | Expected Output |
|---|---|
| `circle(1)` | `(3.14, 6.28)` |
| `circle(5)` | `(78.54, 31.42)` |
| `circle(0)` | `(0.0, 0.0)` |

```python
def circle(radius):
    pi = 3.14159
    return (round(pi * radius ** 2, 2), round(2 * pi * radius, 2)) 
```

## Exercise 2 вАФ Even or Odd

Use the **modulus operator `%`** to determine if a number is even or odd. Return `'even'` or `'odd'`.

вЪ†пЄП **Constraint:** No `if/else` allowed вАФ figure out how to use `%` alone to produce the answer.

| Input | Expected Output |
|---|---|
| `even_or_odd(4)` | `'even'` |
| `even_or_odd(7)` | `'odd'` |
| `even_or_odd(0)` | `'even'` |
| `even_or_odd(-3)` | `'odd'` |

```python
def even_or_odd(n):
    return "odd" if n % 2 == 1 else "even"
```

## Mini-Project вАФ Simple Calculator

Build a calculator that takes two numbers and an operator string.

`op` will be one of: `'+'`, `'-'`, `'*'`, `'/'`, `'//'`, `'%'`, `'**'`

вЪ†пЄП If `op` is `'/'` or `'//'` and `b` is `0`, return `'Error: division by zero'`

| Input | Expected Output |
|---|---|
| `calculator(10, '+', 5)` | `15` |
| `calculator(2, '**', 8)` | `256` |
| `calculator(5, '/', 0)` | `'Error: division by zero'` |

```python
def calculator(a, op, b):
    if (op == "//" or op == "/") and b == 0:
        return "Error: division by zero"
    elif(op == "+"):
        return a + b
    elif(op == "-"):
        return a - b
    elif(op == "/"):
        return a / b
    elif(op == "//"):
        return a // b
    elif (op == "%"):
        return a % b
    elif(op == "*"):
        return a * b
    elif(op == "**"):
        return a ** b
    
```
