## Exercise 1 — List Stats

Python has powerful built-in functions that work on collections. Your job is to use `len()`, `min()`, `max()`, and `sum()` to analyze a list of numbers.

Return a dictionary with these keys: `'length'`, `'min'`, `'max'`, `'sum'`

| Input | Expected Output |
|---|---|
| `[3, 1, 4, 1, 5, 9]` | `{'length': 6, 'min': 1, 'max': 9, 'sum': 23}` |
| `[10, 20, 30]` | `{'length': 3, 'min': 10, 'max': 30, 'sum': 60}` |

```python
def list_stats(numbers):
    return {"length": len(numbers), "min": min(numbers), "max": max(numbers), "sum": sum(numbers)}
```

## Mini-Project — Unit Converter

Write a function that converts a value from one unit to another based on the `unit` string passed in.

**Supported conversions:**

| unit | formula |
|---|---|
| `'kg_to_lbs'` | `value * 2.205` |
| `'lbs_to_kg'` | `value / 2.205` |
| `'km_to_miles'` | `value * 0.621` |
| `'miles_to_km'` | `value / 0.621` |
| `'c_to_f'` | `(value * 9/5) + 32` |
| `'f_to_c'` | `(value - 32) * 5/9` |

Return result **rounded to 2 decimal places**.

```python
def convert(value, unit):
    if (unit == "kg_to_lbs"):
        return round(value * 2.205, 2)
    elif (unit == "lbs_to_kg"): 
        return value / 2.205  
    elif (unit == "km_to_miles"):
        return value * 0.621
    elif (unit == "miles_to_km"):
        return value / 0.621
    elif (unit == "c_to_f"):
        return (value * 9/5) + 32
    elif (unit == "f_to_c"):
        return (value - 32) * 5/9
```

```python
# TEST CELL — do not modify
def _check(label, got, expected):
    if got == expected:
        print(f'  ✔ APPROVED — {label}')
    else:
        print(f'  ✘ FAILED   — {label}')
        print(f'             got:      {got!r}')
        print(f'             expected: {expected!r}')

print('\n── Mini-Project: convert ──')
_check('1 kg_to_lbs',    convert(1,   'kg_to_lbs'),    2.21)
_check('100 c_to_f',     convert(100, 'c_to_f'),       212.0)
_check('0 c_to_f',       convert(0,   'c_to_f'),       32.0)
_check('10 km_to_miles', convert(10,  'km_to_miles'),  6.21)
```
