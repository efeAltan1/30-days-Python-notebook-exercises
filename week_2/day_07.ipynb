## Exercise 1 — Set Operations

Return a dictionary with all five set operations between sets `a` and `b`:

| Key | Operation | Meaning |
|---|---|---|
| `'union'` | `a \| b` | all elements from both |
| `'intersection'` | `a & b` | elements in both |
| `'diff_a'` | `a - b` | in `a` but NOT in `b` |
| `'diff_b'` | `b - a` | in `b` but NOT in `a` |
| `'symmetric'` | `a ^ b` | in either, but NOT in both |

**Test case:**
```python
set_ops({1,2,3,4}, {3,4,5,6})
→ {'union': {1,2,3,4,5,6}, 'intersection': {3,4},
   'diff_a': {1,2}, 'diff_b': {5,6}, 'symmetric': {1,2,5,6}}
```

````python
def set_ops(a, b):
    return {

        "union": a | b,
        "intersection": a & b,
        "diff_a": a - b,
        "diff_b": b - a,
        "symmetric": a ^ b

    }
````

````python
# TEST CELL — do not modify
def _check(label, got, expected):
    if got == expected:
        print(f'  ✔ APPROVED — {label}')
    else:
        print(f'  ✘ FAILED   — {label}')
        print(f'             got:      {got!r}')
        print(f'             expected: {expected!r}')

print('\n── Exercise 1: set_ops ──')
result = set_ops({1,2,3,4}, {3,4,5,6})
_check('union',        result.get('union'),        {1,2,3,4,5,6})
_check('intersection', result.get('intersection'), {3,4})
_check('diff_a',       result.get('diff_a'),       {1,2})
_check('diff_b',       result.get('diff_b'),       {5,6})
_check('symmetric',    result.get('symmetric'),    {1,2,5,6})
````

## Exercise 2 - Find Duplicates

Given a list, return a set of items that appear more than once.

| Input | Expected Output |
|---|---|
| [1,2,3,2,4,3,5] | {2, 3} |
| ['a','b','a','c'] | {'a'} |
| [1,2,3] | set() |

````python
def find_duplicates(items):
    dupes = set()
    for d in items:
        if items.count(d) > 1:
            dupes.add(d)
    return dupes
````

````python
def _check(label, got, expected):
    if got == expected:
        print(f'  APPROVED - {label}')
    else:
        print(f'  FAILED   - {label}')
        print(f'             got:      {got!r}')
        print(f'             expected: {expected!r}')

print('\n-- Exercise 2: find_duplicates --')
_check('[1,2,3,2,4,3,5]',    find_duplicates([1,2,3,2,4,3,5]),   {2,3})
_check('[a,b,a,c]',           find_duplicates(['a','b','a','c']), {'a'})
_check('[1,2,3] no dups',     find_duplicates([1,2,3]),           set())
````

## Mini-Project - Unique Word Analyser

Given two strings text1 and text2, return a dictionary:

| Key | Value |
|---|---|
| 'common' | words in both texts (set) |
| 'only_in_first' | words only in text1 (set) |
| 'only_in_second' | words only in text2 (set) |
| 'total_unique' | count of all unique words across both |

Hint: convert each text to a set of lowercase words using .lower().split()

````python
def word_analysis(text1, text2):
    conv_text_1=set(text1.lower().split())
    conv_text_2=set(text2.lower().split())

    return{

        "common": conv_text_1 & conv_text_2,
        "only_in_first": conv_text_1 - conv_text_2,
        "only_in_second": conv_text_2 - conv_text_1,
        "total_unique": len(conv_text_1 | conv_text_2)

        
    }
````

````python
def _check(label, got, expected):
    if got == expected:
        print(f'  APPROVED - {label}')
    else:
        print(f'  FAILED   - {label}')
        print(f'             got:      {got!r}')
        print(f'             expected: {expected!r}')

print('\n-- Mini-Project: word_analysis --')
r = word_analysis('the cat sat on the mat', 'the dog sat on the log')
_check('common',         r.get('common'),         {'the','sat','on'})
_check('only_in_first',  r.get('only_in_first'),  {'cat','mat'})
_check('only_in_second', r.get('only_in_second'), {'dog','log'})
_check('total_unique',   r.get('total_unique'),   7)
````
